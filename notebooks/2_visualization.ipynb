```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import constants as const
from scipy.fft import fft, fftfreq, fftshift
```

```python
def plot_raw(df, figsize=(10, 5)):
    len_pos = len(const.POS)
    len_type = len(const.TYPES)
    total = len_pos * len_type
    for i in range(len_pos):
        for j in range(len_type):
            cols = [f"{const.POS[i]}_{const.SENSORS[3*j+k]}" for k in range(3)]
            plt.figure(figsize=figsize)
            plt.plot(df[cols], label=cols)
            plt.title(f"{const.POS[i]}_{const.TYPES[j]}")
            plt.legend()
            plt.show()
            
def plot_freq(df, figsize=(20, 10)):
    n = const.NUM_SAMPLES//2
    plt.figure(figsize=figsize)
    xf = fftshift(fftfreq(const.NUM_SAMPLES, 1/const.FREQ))
    xplot = xf[n:]
    for col in const.COLUMNS:
        yf = fftshift(fft(df[col].values))
        yplot = 1.0/const.NUM_SAMPLES * np.abs(yf)[n:]
        plt.plot(xplot, yplot, label=col)
    plt.legend()
    plt.show()
```

<h4>Stair up</h4>

```python
stair_up_window = f"{const.WINDOWS_FOLDER}/T0_Id019341.actLabel/00000016.csv"
df = pd.read_csv(stair_up_window)
lbl = df.Label.unique()[0]
print(const.MAPPING_LABELS[lbl])
```

```python
plot_raw(df)
```

```python
plot_freq(df)
```
